```python
# dominios

import math
import matplotlib.pyplot as plt
import sympy as sp
sp.var("x")
sp.var("y")
def f(x):
    return (x**2)-(4*x)+2
sp.solve(f(x),x)
print("Dom: x incluye R, no incluye "+ str(sp.solve(f(x),x)))
def g(x):
    return sp.sin(3*x)
sp.solve(g(x),x)
print("Dom: x incluye R, no incluye "+ str(sp.solve(g(x),x)))
def h(x):
    return sp.exp(2*x)-3
sp.solve(h(x),x)
print("Dom: x incluye R, no incluye "+ str(sp.solve(h(x),x)))
def k(x):
    return sp.sqrt((x**2)-4)
sp.solve(k(x),x)
print("Dom: x incluye R, no incluye "+ str(sp.solve(k(x),x)))
def p(x):
    return sp.cos(x+(math.pi/4))
sp.solve(p(x),x)
print("Dom: x incluye R, no incluye "+ str(sp.solve(p(x),x)))

```

```python
#graph

import numpy as np
x = np.linspace(-1,5,1000)
plt.figure(figsize=(5,5))
plt.plot(x,f(x),label="function1")
plt.plot(x,np.sin(3*x), label="function2")
plt.grid()
plt.title("functions")
plt.legend()
```
