### Ejercicio 1

Se quiere producir comida para gatos de la manera más barata, no obstante se debe también asegurar que se cumplan los datos requeridos de analisis nutricional. Por lo que se quiere variar la cantidad de cada ingrediente para cumplir con los estandares nutricionales. Los requisitos que se tienen es que en 100 gramos, se deben tener por lo menos 9 gramos de proteína y  7 gramos de grasa. Así mismo, no se debe tener más de 2.5 gramos de fibra y 0.5 gramos de sal.

Los costos de producto son:

1. IDENTIFIQUE CRITERIO
   
   * Se desea un MIN de COSTO
    
2. Variables de control
   * gramos de pollo --> $x_P$
   * gramos de carne --> $x_Ca$
   * gramos de cordero --> $x_Co$
   * gramos de arroz --> $x_A$
   * gramos de trigo --> $x_T$
   * gramos de gel --> $x_G$


3. FUNCIÓN DE COSTO

   $f(x_P,x_Ca,x_Co,x_A,x_T) = 0.012\frac{\$}{g}x_P + 0.009\frac{\$}{g} x_Ca+ 0.010 \frac{\$}{g}x_Co + 0.003\frac{\$}{g}x_T + 0.005\frac{\$}{g}x_T+0.0\frac{\$}{g}x_G$ 
4. INFORMACIÓN ADICIONAL
   * POLLO
       * 10% proteína
       * 8% grasa
       * 0.1% fibra
       * 0.2% sal
    * CARNE
       * 20% proteína
       * 10% grasa
       * 0.5% fibra
       * 0.5% sal
    * CORDERO
       * 10% proteína
       * 8% grasa
       * 0.1% fibra
       * 0.2% sal
    * ARROZ
       * 10% proteína
       * 8% grasa
       * 0.1% fibra
       * 0.2% sal
    * TRIGO
       * 10% proteína
       * 8% grasa
       * 0.1% fibra
       * 0.2% sal
    * GEL
      * 0 prote
      * 0 grasa
      * 0 fibra
      * 0 sal

    * PRODUCTO FINAL (RESTRICCIONES)
      * por lo menos 9 gramos de proteína
      * por lo menos 7 gramos de grasa
      * menor o igual a 2.5 gramos de fibra
      * menor o igual a 0.5 gramos de sal

$$\min_{x_P,x_Ca,x_Co,x_A,x_T} 0.012x_P + 0.009x_Ca + 0.010x_Co + 0.003x_T + 0.005x_T + 0.00X_G$$
$$\text {s.t.} $$
$$ -0.10 x_P + -0.20 x_Ca + -0.15 x_Co + -0 x_A + -0.04 x_T + -0 x_G \leq -9 $$
$$ -0.08 x_P + -0.10 x_Ca + -0.11 x_Co + -0.01 x_A + -0.01 x_T + -0 x_G \leq -7 $$
$$ 0.01 x_P + 0.05 x_Ca + 0.05 x_Co + 0.10 x_A + 0.15 x_T + 0 x_G \leq 2.5$$
$$ 0.002 x_P + 0.005 x_Ca + 0.007 x_Co + 0.002 x_A + 0.008 x_T + 0 x_G \leq 0.5$$
$$  x_P +  x_Ca +  x_Co +  x_A +  x_T +  x_G = 100$$

    

```python
import numpy as np
import scipy.optimize as opt

c = ([
    0.012,
    0.009,
    0.010,
    0.003,
    0.005,
    0.00
])

Aub = np.array([
    [ -0.10,  -0.20,  -0.15,     -0,  -0.04,  0],
    [ -0.08,  -0.10,  -0.11,  -0.10,  -0.10,  0],
    [  0.01,   0.05,   0.05,   0.10,   0.15,  0],
    [ 0.002,  0.005,  0.007,  0.002,  0.008,  0], 
])

Aeq = np.array([
    [1,1,1,1,1,1]
])
bub = np.array([
    -9,
    -7,
    2.5,
    0.5,
])
beq = np.array([
    100
])
sol = opt.linprog(c,Aub,bub,Aeq,beq)
sol
```

### Ejercicio 2.1
Una multinacional farmacéutica desea fabricar un compuesto nutritivo a base de dos productos A y B.  El producto A contiene  $30\%$ de proteínas,  un $1\%$ de grasas y un  $10\%$ de azúcares.  El producto B  contiene  un  $5\%$  de proteínas,  un  $7\%$ de grasas  y un  $10\%$  de  azúcares.

El compuesto debe de tener,  al menos, $25g$ de proteínas,  $6g$ de grasas y $30g$ de azúcares.  El coste del producto A es de $0.6$ u.m./g  y el de B es de $0.2$  u.m./g

Se desea encontrar la cantidad en gramos de cada producto para que el coste total sea mínimo.

Formular el problema de decidir cuánto hacer de cada producto como un problema de programación lineal.


1. Se busca...
   * MIN costos
2. Variables
   * proporción de A
   * proporción de B
3. función de costo
   $$f(X_A,X_B) = 0.6X_a + 0.2X_B$$
4. RESTRICCIONES
$$ \min_{X_A,X_B} = 0.6X_A + 0.2X_B$$
$$ \text {s.t.} $$
$$ -0.3X_A - 0.05X_B \leq -25$$
$$ -0.1X_A - 0.07X_B \leq -6$$
$$ -0.1X_A - 0.1X_B \leq -30$$

```python
import numpy as np

c = np.array([0.6,0.2])

Aub = np.array([
    [-0.3,-0.05],
    [-0.1, -0.07],
    [-0.1, -0.1]
])
bub = np.array([-25,-6,-30])
import scipy.optimize as opt

sol= opt.linprog(c,Aub,bub)
sol
```

```python
c @ sol.x
```
