 Utiliza sympy y encuentra los valores de las incgnitas que hagan que los siguientes sistemas de ecuaciones sean verdaderos:
 

```python
import sympy as sp
sp.var("x")
sp.var("y")
from sympy import*
x,y=symbols("x y")
#defino ecuaciones
eq1 = x + y -10
eq2 = x - y -2

#resuleva una variable en una ecuacion
x_eq1 = sp.solve(eq1,x)[0]
x_eq1

#sustituya en segunda ecuacion
y_eq2 = eq2.subs(x,x_eq1)

#resuelva una segunda ecuacion
y_sol = sp.solve(y_eq2,y)[0]

#sustituye valor de y para valor de x
x_sol = x_eq1.subs(y,y_sol)

x_sol, y_sol
```

```python
import sympy as sp
sp.var("x")
sp.var("y")

#defino ecuaciones
eq1 = ((x**2)-(2*y))-16
eq2 = x - y -3

#resuleva una variable en una ecuacion
x_eq1 = sp.solve(eq1,x)[0]

x2_eq1=sp.solve(eq1,x)[1]

eq2enY = sp.solve(eq2,y)[0]

PrimSUS= x_eq1.subs(y,eq2enY)
SegSUS= x2_eq1.subs(y,eq2enY)

x_sol=sp.solve(PrimSUS,x)
x2_sol=sp.solve(SegSUS,x)

x_sol,x2_sol
```

```python
import sympy as sp
sp.var("x")
sp.var("y")
sp.var("z")

#defino ecuaciones
eq1 = x + y - z -10
eq2 = x - y - 2
eq3 = y + z + 2

#resuleva una variable en una ecuacion
y_eq2 = sp.solve(eq2,y)[0]

z_eq3 = sp.solve(eq3,z)[0]

susXenZ = z_eq3.subs(y,y_eq2)
susYenX = eq1.subs(y,y_eq2)
susZenX = susYenX.subs(z,susXenZ)
x_sol= sp.solve(susZenX,x)[0]

yeq=eq2.subs(x,x_sol)
y_sol=sp.solve(yeq,y)[0]

zeq=eq3.subs(y,y_sol)
z_sol=sp.solve(zeq,z)[0]

x_sol,y_sol,z_sol
```
