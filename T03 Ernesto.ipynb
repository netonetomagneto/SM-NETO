```python
import sympy as sp
sp.var("x")
sp.var("y")
# la base se conideró como "2x" y los lados iguales como "y" cada uno
perimetro = 2*x + 2*y - 20
#perímetro en función de "y" en términos de "x", resultando en una "y" sustituible.
periY = sp.solve(perimetro,y)[0]

#sustituir la solución de periY en y
h= ((periY**2-x**2)**(1/2))
areaT = x*h

#derivar función ed área con respecto de x e igualar a 0 para encontrar el valor crítico de x

difx = sp.diff(areaT, x)
Xmax = sp.solve(difx,x)[0]

#obtener segunda derivada y sustituir x1 para verificar que es menor a 0, indicando un valor máximo.

difx2 = sp.diff(difx,x)
x2 = difx2.subs(x,Xmax)

#sustituir el valor max de x en la función de "y"

Ymax = periY.subs(x,Xmax)
Ymax,Xmax
base = Xmax*2
lado = Ymax
print("valor máximo base = " + str(base))
print("valor máximo cada lado igual = " +str(lado))
```

Disponemos de una barra de aluminio de 6 metros para consturi una portería de fútbol en un parque, Es unaportería simple que tendrá sólo dos postes y el travesaño:

Si queremos que el área de la portería sea máxima, ¿Cuánto deben medir los postes y el travesaño?

```python
import sympy as sp
sp.var("x")
sp.var("y")
# "y" postes y "2x" travesaño. obtenemos la función y respecto a x para ajustar

barra = (2*x)+(2*y)-6
Fy = sp.solve(barra,y)[0]

area = x*Fy
difx = sp.diff(area,x)
Xmax = sp.solve(difx,x)[0]
dif2x = sp.diff(area,x,2)

# segunda derivada menor a 0 con Xmax sustituido (MAX < 0)
MAX = dif2x.subs(x,Xmax)
MAX

Poste = Fy.subs(x,Xmax)
Trave = 2*Xmax
print("Valor max travesaño = " + str(Trave))
print("Valor max de cada poste = " + str(Poste))
```

**Resistencia de una viga**. Una viga de madera tiene una sección transversal rectangular de altura _h_  y ancho _w_. La resistencia **S** de la viga es directamente proporcional al ancho y al cuadrado de su altura (**S = kwh^2**, donde _**k**_ es una constante de proporcionalidad). 

¿Cuáles son las dimensiones de la viga más fuerte que puede cortarse a partir de un leño redondo de **24** pulgadas de diámetro? 


```python
import sympy as sp
sp.var("w")
sp.var("h")

#limitante de 24 pulgadas.
TriRectLim = h**2+w**2-576

#No puede existir una medida negativa, por lo que se elimina la raiz negativa y se toma la positiva ([1]) 
W = sp.solve(TriRectLim,w)[1]

S = W*h**2
Sh = sp.diff(S,h)
Hmax = sp.solve(Sh,h)[2]
# se eliminan la raiz negativa y el 0.

#Verificamos (MAX < 0)
Sh2 = sp.diff(S,h,2)
MAX= Sh2.subs(h,Hmax)

#obtenemos valor de w, sustituyendo
Wmax = W.subs(h,Hmax)
Wmax
Smax = Wmax * Hmax**2
Smax
print("Valor max de S = k("+ str(Smax)+")")
```

**Iluminacion** Una fuente luminosa se localiza sobre el centro de una mesa circular de 4 pies de diámetro. Encontrar la altura de **_h_** de la fuente luminosa de modo tal que la iluminación **_I_** en el perímetro de la mesa sea máxima si **_I=k(sina)/s^2_**, donde s es la altura oblicua, **_a_** es el angulo al cual la luz incide sobre la mesa y **_k_** es una constante (El valor de esta última constante es irrelevante para el problema, ya que no se pide el valor de la iluminación)

```python
import sympy as sp
sp.var("h")
sp.var("s")

#Definimos S en terminos de h para poder derivar y sustituimos en I
s = (h**2+4)**(1/2)

#definimos sin(a) como OP/HIP (h/s) y sustituimos

I = h/(s**3)
Ih = sp.diff(I,h)
Hmax = sp.solve(Ih,h)[1]
#tomamos el valor positivo [1]

#Verificamos (MAX < 0)
Ih2 = sp.diff(I,h,2)
MAX = Ih2.subs(h,Hmax)

print("h max = "+ str(Hmax))
```
