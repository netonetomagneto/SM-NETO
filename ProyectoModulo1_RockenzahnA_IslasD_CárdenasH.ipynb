## **INTRODUCCIÓN**  

En los últimos años la nanotecnología ha mostrado avances impresionantes en diversas áreas, especialmente en el de la óptica. Al estudiar, analizar y manipular las propiedades de la luz se han logrado hallazgos como la fibra óptica o las radiografías. Pero hoy en día, esta área de estudio presenta una nueva oportunidad con los “metamateriales”.  
Para entender el proyecto a continuación y sus propósitos, se explicarán dos conceptos importantes; los *metamateriales* y el *índice de refracción* de la luz en ellos.    
Los metamateriales son materiales artificiales que no se encuentran en la naturaleza y presentan propiedades electromagnéticas inusuales, pueden por ejemplo llegar a índices de refracción negativos o cercanos a cero.   
El índice de refracción se define como el cociente de la velocidad de la luz cuando pasa a través de dos medios, o en otras palabras, define la rapidez con la que un haz de luz viaja a través de los medios, y esta relación se describe mediante la fórmula:   
$$n=c/v$$   
Donde:  
n es el índice de refracción, c es la velocidad de la luz en un vacío o aire, v es al velocidad de la luz en los medios (por ejemplo agua, aceite de oliva etc...)   
Entonces ¿qué hace realmente un metamaterial? ¿Por qué es tan importante?  
Normalmente los índices de refracción en los materiales son positivos, pero en los metamateriales, a determinadas longitudes de onda, se puede hacer que este índice sea cercano a cero.   
Como el índice de refracción describe la capacidad de un material para ralentizar la luz a medida que pasa a través de él. Cuando el índice de refracción es cercano a cero en un metamaterial:  
- La velocidad de fase de la luz se vuelve extremadamente alta, es decir, la onda parece desplazarse sin experimentar un cambio significativo en su fase.   
- Esto significa que todas las partes de la onda de luz (la cresta, el valle, etc.) parecen moverse al mismo ritmo, casi "sin dispersarse", como normalmente lo harían en otros medios.  
Si la luz puede propagarse a través de un objeto sin cambiar de fase ni dirección (debido a su índice de refracción cero), ese objeto podría "desaparecer" para el observador. No reflejaría ni refractaría luz de la manera convencional, haciéndolo invisible bajo ciertas condiciones.    
Aunque aún está en desarrollo, los materiales ENZ podrían ser usados para desarrollar capas de invisibilidad, que desviaran las ondas de luz sin ser detectadas.  
Si creamos una tela con metamateriales que cubra una manzana a la longitud de onda visible, esta desaparecerá de nuestra vista y veremos lo que hay al otro lado. En un futuro se podrá aplicar para edificios o personas.

# MODELO DIOXIDO DE SILICIO

```python
import pandas as pd
import numpy as np

datos = pd.read_csv('Malitson.csv')

datos = datos[(datos['wl'] >= 0.3) & (datos['wl'] <= 1)]

datos['wl_nm'] = datos['wl'] * 1000

rangos = [
    (300, 400), (400, 500), (500, 600), (600, 700),
    (700, 800), (800, 900), (900, 1000)
]

def regresion_lineal(x, y):
    coef = np.polyfit(x, y, 1)
    return coef[0], coef[1]  

valores_esperados = {
    (300, 400): {'pendiente': -0.00017, 'intercepto': 1.538},
    (400, 500): {'pendiente': -0.00007, 'intercepto': 1.500},
    (500, 600): {'pendiente': -0.00004, 'intercepto': 1.483},
    (600, 700): {'pendiente': -0.00003, 'intercepto': 1.474},
    (700, 800): {'pendiente': -0.00002, 'intercepto': 1.469},
    (800, 900): {'pendiente': -0.000015, 'intercepto': 1.466},
    (900, 1000): {'pendiente': -0.000013, 'intercepto': 1.464},
}

errores = []

for inicio, fin in rangos:
    datos_rango = datos[(datos['wl_nm'] >= inicio) & (datos['wl_nm'] < fin)]
    if len(datos_rango) > 1:
        pendiente, intercepto = regresion_lineal(datos_rango['wl_nm'], datos_rango['n'])
        
        esperado_pendiente = valores_esperados[(inicio, fin)]['pendiente']
        esperado_intercepto = valores_esperados[(inicio, fin)]['intercepto']
        
        error_absoluto_pendiente = abs(pendiente - esperado_pendiente)
        error_absoluto_intercepto = abs(intercepto - esperado_intercepto)
        
        error_relativo_pendiente = (error_absoluto_pendiente / abs(esperado_pendiente)) * 100
        error_relativo_intercepto = (error_absoluto_intercepto / abs(esperado_intercepto)) * 100
        
        errores.append({
            'rango': (inicio, fin),
            'error_absoluto_pendiente': error_absoluto_pendiente,
            'error_relativo_pendiente': error_relativo_pendiente,
            'error_absoluto_intercepto': error_absoluto_intercepto,
            'error_relativo_intercepto': error_relativo_intercepto
        })

for error in errores:
    print(f"Rango {error['rango']} nm:")
    print(f"  Error absoluto pendiente: {error['error_absoluto_pendiente']:.6f}")
    print(f"  Error relativo pendiente: {error['error_relativo_pendiente']:.2f}%")
    print(f"  Error absoluto intercepto: {error['error_absoluto_intercepto']:.6f}")
    print(f"  Error relativo intercepto: {error['error_relativo_intercepto']:.2f}%")
```
